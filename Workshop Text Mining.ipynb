```python
#stemming
stemmer = nltk.PorterStemmer()
verbs = ['calling', 'called', 'appears', 'appear', 'appeared']
stems = []
for verb in verbs :
    stemmed_verb = stemmer.stem(verb)
    stems.append(stemmed_verb)
    
stems
#set(stems)
#sorted (set(stems))
```

```python
nltk.download ('inaugural')
#insugural corpus 
from nltk.corpus import inaugural
inaugural.fileids()[:2]

cfd = nltk.ConditionalFreqDist(
    [(fileid[:4],target) for fileid in inaugural.fileids() for w in inaugural.words(fileid) for target in 
    ['america', 'war'] if w.lower().startswith(target)])
cfd.plot()
```
